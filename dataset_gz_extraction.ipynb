```python
import os
import gzip
import shutil

# Set the root directory
root_dir = "./filtered_balck_box_datasets"

# Traverse all subdirectories
for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        # Check if the file has a .tsv.gz extension
        if file.endswith(".tsv.gz"):
            # Construct the full file paths
            gz_file_path = os.path.join(subdir, file)
            tsv_file_path = os.path.join(subdir, file[:-3])  # Remove ".gz" extension
            
            # Open the .gz file and extract the contents
            with gzip.open(gz_file_path, 'rb') as gz_file:
                with open(tsv_file_path, 'wb') as tsv_file:
                    shutil.copyfileobj(gz_file, tsv_file)
            
            print(f"Extracted: {gz_file_path}")
```
